# AutoEncoder

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input
from sklearn.datasets import load_iris
```

```python
data = load_iris().data
```

```python
input_layer = Input(shape=(4,))

encode_layer1 = Dense(8,activation='relu')(input_layer)
encode_layer2 = Dense(16,activation='relu')(encode_layer1)

middle_layer = Dense(8,activation='relu')(encode_layer2)

decode_layer1 = Dense(16,activation='relu')(middle_layer)
decode_layer2 = Dense(8,activation='relu')(decode_layer1)

output_layer = Dense(4,activation='linear')(decode_layer2)
```

```python
auto_encoder = Model(input_layer,output_layer)
```

```python
auto_encoder.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
auto_encoder.fit(data,data, epochs=300,batch_size=40)
```

```python
decoder_input = Input(shape=(8,))
decoder_layer11 = auto_encoder.layers[4](decoder_input)
decoder_layer12 = auto_encoder.layers[5](decoder_layer11)
output_layer1 = auto_encoder.layers[6](decoder_layer12)

decoder = Model(decoder_input, output_layer1)
```

# Convolutional Neural Network

```python
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train)
```

```python
model = Sequential()

model.add(Conv2D(input_shape=(32,32,3),kernel_size=(3,3),filters=4))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(kernel_size=(2,2),filters=4))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(10,activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=32)
```
