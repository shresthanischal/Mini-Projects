```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import normalize, to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
data = pd.read_csv('D:/DW/startups.csv')
```

```python
data.head()
```

```python
x = data.iloc[:,:-1]
y = data['Profit']
```

```python
x.head()
```

```python
for col in x.columns:
    if col == 'State':
        x[col] = LabelEncoder().fit_transform(x.loc[:,[col]])
    else:
        x[col] = StandardScaler().fit_transform(x.loc[:,[col]])
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
import tensorflow as tf
```

```python
rms = tf.keras.metrics.RootMeanSquaredError()
```

```python
model = Sequential()

model.add(Dense(4,input_dim=4,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1,activation='linear'))

model.compile(optimizer='adam',loss='mse',metrics=[rms])
```

```python
model.fit(x_train,y_train,epochs=1000,batch_size=5)
```

```python
temp = model.predict(x_test.iloc[1:,:])
```

```python
data =  pd.read_csv('D:/DW/tips.csv')
```

```python
data.head()
```

```python
x = data.drop(['tip'],axis=1)
y = data['tip']
```

```python
x.head()
```

```python
cat = x.select_dtypes(include=[object])
num = x.select_dtypes(exclude=[object])

for i in cat.columns:
    x[i] = LabelEncoder().fit_transform(x.loc[:,[i]])
    
for i in num.columns:
    x[i] = StandardScaler().fit_transform(x.loc[:,[i]])
```

```python
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
model1 = Sequential()

model1.add(Dense(4,input_dim=6,activation='relu'))
model1.add(Dense(8,activation='relu'))
model1.add(Dense(16,activation='relu'))
model1.add(Dense(32,activation='relu'))
model1.add(Dense(64,activation='relu'))
model1.add(Dense(32,activation='relu'))
model1.add(Dense(16,activation='relu'))
model1.add(Dense(8,activation='relu'))
model1.add(Dense(4,activation='relu'))
model1.add(Dense(1,activation='linear'))

model1.compile(optimizer='adam',loss='mse',metrics=[rms])
```

```python
model1.fit(x_train,y_train,epochs=1000,batch_size=32)
```
